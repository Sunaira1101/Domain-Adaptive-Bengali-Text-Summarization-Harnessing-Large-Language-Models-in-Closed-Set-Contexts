```python
# Function to generate predictions
def generate_summary(input_text, model, tokenizer, max_length=150):
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, max_length=512)
    outputs = model.generate(
        inputs["input_ids"],
        max_length=max_length,
        num_beams=4,
        early_stopping=True
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
from rouge import Rouge
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from bert_score import score

def evaluate_model(predictions, references, calculate_bert_score=False, bert_score_model='bert-base-uncased'):

    # Initialize metrics
    rouge = Rouge()
    smoothing_function = SmoothingFunction().method1

    # Metric accumulators
    rouge_1_f1, rouge_2_f1, rouge_l_f1 = 0, 0, 0
    bleu_scores = []

    # Calculate ROUGE and BLEU scores
    for pred, ref in zip(predictions, references):
        # ROUGE scores
        rouge_scores = rouge.get_scores(pred, ref, avg=True)
        rouge_1_f1 += rouge_scores['rouge-1']['f']
        rouge_2_f1 += rouge_scores['rouge-2']['f']
        rouge_l_f1 += rouge_scores['rouge-l']['f']

        # BLEU score
        pred_tokens = pred.split()
        ref_tokens = ref.split()
        bleu_score = sentence_bleu([ref_tokens], pred_tokens, smoothing_function=smoothing_function)
        bleu_scores.append(bleu_score)

    # Calculate average scores
    n = len(predictions)
    rouge_1_f1 /= n
    rouge_2_f1 /= n
    rouge_l_f1 /= n
    avg_bleu = sum(bleu_scores) / n

    # Initialize results
    results = {
        "ROUGE-1 F1": rouge_1_f1,
        "ROUGE-2 F1": rouge_2_f1,
        "ROUGE-L F1": rouge_l_f1,
        "BLEU": avg_bleu,
    }

    # Optionally calculate BERTScore
    if calculate_bert_score:
        bert_p, bert_r, bert_f1 = score(predictions, references, lang="en", model_type=bert_score_model)
        results["BERTScore Precision"] = bert_p.mean().item()
        results["BERTScore Recall"] = bert_r.mean().item()
        results["BERTScore F1"] = bert_f1.mean().item()

    return results
```

```python
# Evaluate with BERTScore
results_with_bert = evaluate_model(zero_shot_predictions_state, references, calculate_bert_score=True, bert_score_model="bert-base-uncased")
print(results_with_bert)
```
